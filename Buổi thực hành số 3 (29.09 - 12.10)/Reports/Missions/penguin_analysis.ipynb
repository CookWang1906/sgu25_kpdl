## **Import thư viện và chuẩn bị dữ liệu phân tích**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

penguins_data = pd.read_csv(r"C:\Users\qhoan\.vscode\Coding\Python\Learning\Khai phá dữ liệu\Lab03_EDA_TranQuocHoang_3123580015\Data\penguins.csv")
penguins_data = penguins_data[['Species','Culmen Length (mm)','Body Mass (g)','Sex']]
print(penguins_data.head(5))
print(penguins_data.shape)
```

## **Phân tích dữ liệu 2 biến dựa vào phương pháp scatterplot**

```python
# case 1: basic
sns.scatterplot(data = penguins_data, x = penguins_data["Culmen Length (mm)"],
y = penguins_data['Body Mass (g)'])
```

```python
# case 2: advanced
plt.figure(figsize= (12,6))
ax = sns.scatterplot(data = penguins_data,
x= penguins_data["Culmen Length (mm)"],
y= penguins_data['Body Mass (g)'])
ax.set_title('Bivariate analysis of Culmen Length and body mass', fontsize = 20)
```

## **Phân tích 2 biến dựa vào bảng crosstab/two-way**

```python
pd.crosstab(index= penguins_data['Species'], columns= penguins_data['Sex'])
```

## **Phân tích 2 biến sử dụng pivot_table**

```python
pd.pivot_table(penguins_data, values='Culmen Length (mm)', index='Species', aggfunc = np.mean)
```

## **Phân tích 2 biến sử dụng pairplot**

```python
import seaborn as sns
sns.pairplot(data = penguins_data)
```
